```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets, linear_model, model_selection
```

```python
# https://scikit-learn.org/stable/datasets/toy_dataset.html#diabetes-dataset

# X is the data matrix (features) and y the data target (here disease progression)
X, y = datasets.load_diabetes(return_X_y=True)
```

```python
features = {
    "age": 0,
    "sex": 1,
    "body mass index": 2,
    "average blood pressure": 3,
    "total serum cholesterol": 4,
    "low-density lipoproteins": 5,
    "high-density lipoproteins": 6,
    "total cholesterol": 7,
    "log of serum triglycerides level": 8,
    "blood sugar level": 9,
}

selected = list(features.keys())[3]
```

```python
# We need to be able to plot data so we keep only the third column, and then we
# reshape the array to make 1 col instead of a list
# reshape(n_rows, n_columns). If one of the parameter is -1, the corresponding dimension is calculated automatically.

X = X[:, features[selected]]
X = X.reshape((-1,1))
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel(selected)
plt.ylabel('Disease Progression')
plt.title(f'A Graph Plot Showing Diabetes Progression Against {selected}')
plt.show()
```
