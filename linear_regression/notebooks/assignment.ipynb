```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn import datasets, linear_model, model_selection
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_linnerud.html#sklearn.datasets.load_linnerud
# X are the features: the 3 exercices -> Chins Situps, Jumps
# Y are the possible data targets ->  Weight, Waist, Pulse
X, y = datasets.load_linnerud(return_X_y =True)
```

```python
exercices = ["Chins", "Situps", "Jumps"]
targets = ["Weight", "Waist", "Pulse"]

selected = ("Situps", "Waist")
```

```python
# First we need to keep only the feature column we want

X = X[:, exercices.index(selected[0])].reshape(-1, 1)

# And the target column we want

y = y[:, targets.index(selected[1])].reshape(-1, 1)
```

```python
# Then we split data into training and testing part

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
# We train our model

model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# We make the predition with the test data

y_pred = model.predict(X_test)
```

```python
# And we plot the data

plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel(selected[0])
plt.ylabel(selected[1])
plt.title(f'A Graph Plot Showing {selected[1]} Progression depending on {selected[0]}')
plt.show()
```
